### remove html tags

```python
import re

def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'', text)
```

```python
data['review'] = data['review'].apply(remove_html_tags)
```

### Remove punctuation

```python
import string

exclude = string.punctuation
exclude
```

```python
def remove_punc(text):
  for char in exclude:
    text = text.replace(char,'')
  return text
```

```python
data
```

```python
def remove_punc1(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
data['review'] = data['review'].apply(remove_punc1)
```

### Removing stop words

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
stopwords.words('english')
```

```python
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

def remove_sw(text):
  new_text = []

  for word in text.split():
    if word in STOPWORDS:
      new_text.append('')
    else:
      new_text.append(word)
  x = new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
text = data['review'][3]
text
```

```python
remove_sw(text)
```

```python
data['review'] = data['review'].apply(remove_sw)
```

### stemming & lemmetization

```python
# stemming is the process of reducing inflection in words to their root form, sometimes the stem word may not be a valid word.
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
text = data['review'][4]
text
```

```python
stem(text)
```

```python
# in lemmatization, it reduce the inflected word in a way that that word belongs to the language.
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations=" ?:!.: "
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
  if word in punctuations:
    sentence_words.remove(word)

sentence_words
print("{0:20} {1:20}".format("Word", "Lemma"))
for word in sentence_words:
  print ("{0:20}{1:20}".format(word, wordnet_lemmatizer.lemmatize(word,pos='v')))
```
