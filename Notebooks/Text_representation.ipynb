# Text Representation techniques

### Bag Of Words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'Text':['Prachi writes code','Code writes code','Prachi reads book','Code reads book'], 'Output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['Text'])
```

```python
# the numbers given beside the word is index

print(cv.vocabulary_)
```

```python
# Converting first sentence into vector

print(bow[0].toarray())
print(bow[1].toarray())
```

### N-Grams

##### In N-grams if n=(2,2), it will create the pair of 2 words from the corpus sequentially and if (1,2) then it will add both single words and the pair of 2 words into the vocabulary

```python
cv1 = CountVectorizer(ngram_range=(2,2))
```

```python
bow1 = cv1.fit_transform(df['Text'])
```

```python
print(cv1.vocabulary_)
```

```python
print(bow1[0].toarray())
```

### Tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf.fit_transform(df['Text']).toarray()
```

```python
print(tfidf.idf_)
```

##### Tf : term frequesncy
 IDF : Inverse document frequecy

- Tf is calculated by dividing frequency of that term in documents by total number of terms in document
- IDf is calculated by taking the log of diving total no. of documents in the corpus by no. of documents containing that term

## Two types of Word2vec

- CBOW (Continuous bag of words): Given context word predict target word

- Skip-grams: Given target word predict context word

#### - if working with small data use CBOW (fast & accurate for small)
#### - if large data then go with skip-gram
