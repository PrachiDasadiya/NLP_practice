```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc1 = nlp(u"I am learning the spacy right now!")
```

```python
doc1.text
```

```python
doc1[0]
```

```python
doc1[1].pos_
```

```python
doc1[3].tag_
```

```python
spacy.explain(doc1[1].tag_)
```

```python
for word in doc1:
  print(word, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
doc2 = nlp("I read a book.")
doc3 = nlp("I have read a book.")
```

```python
for word in doc2:
  print(word, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

```python
for word in doc3:
  print(word, "----->", word.pos_, word.tag_, spacy.explain(word.tag_))
```

### Visualization

```python
from spacy import displacy
```

```python
doc6 = nlp(u"The quick brown fox jumped over the lazy dog.")
```

```python
displacy.render(doc6,style='dep')
```
